The test of Jaccbian matrix and Hessian matrix in 3D points matching problem. 

```python
import sympy
from sympy import diff, Matrix, Array,symbols
from IPython.display import Math


a1,a2,a3 = symbols('a1,a2,a3')
b1,b2,b3 = symbols('b1,b2,b3')
a = Matrix([a1,a2,a3,1])
b = Matrix([b1,b2,b3,1])
I=Matrix.eye(4)
t11,t12,t13,t14,t21,t22,t23,t24,t31,t32,t33,t34 = symbols('t11,t12,t13,t14,t21,t22,t23,t24,t31,t32,t33,t34')
T0 = Matrix([[t11,t12,t13,t14],
                   [t21,t22,t23,t24],
                   [t31,t32,t33,t34],[0,0,0,1]])
R0 = Matrix([[t11,t12,t13],
                   [t21,t22,t23],
                   [t31,t32,t33]])
# the parameters of SE3
x1,x2,x3,x4,x5,x6 = symbols('x1,x2,x3,x4,x5,x6')
x = Matrix([x1,x2,x3,x4,x5,x6])

s11,s12,s13,s22,s33,s23 = symbols('s11,s12,s13,s22,s33,s23')

# Sigma is a diagonal matrix
Sigma = Matrix([[s11,s12,s13],
                [s12,s22,s23],
                [s13,s23,s33]])
def skew(x):
    return Matrix([[0,-x[2],x[1]],[x[2],0,-x[0]],[-x[1],x[0],0]])

def hat(x):
    return Matrix([[0,-x[5],x[4],x[0]],[x[5],0,-x[3],x[1]],[-x[4],x[3],0,x[2]],[0,0,0,0]])

expx = I + hat(x)
```

The r is the residual vector, and x is the parameters of r.  
* $x \in \Bbb R_n $
* $r \in \Bbb R_m $ 

The Jacabian of r is a $m \times n$ matrix. 

$$
\dot{r} = J_r = 
\begin{bmatrix}
 \frac{\partial{r_1}}{\partial{x_1}}    & \cdots & \frac{\partial{r_1}}{\partial{x_n}}       \\  
 \vdots & \ddots & \vdots\\  
 \frac{\partial{r_m}}{\partial{x_0}}    & \cdots & \frac{\partial{r_m}}{\partial{x_n}}       \\  
\end{bmatrix}$$
$$



The Hessian of r is a $ m \times n \times n$ Tensor. 

$$
\ddot{r_i} = H_{r_i} = 
\begin{bmatrix}
 \frac{\partial^2{r_i}}{\partial{x_1}\partial{x_1}} & \frac{\partial^2{r_i}}{\partial{x_1}\partial{x_2}}  & \cdots & \frac{\partial{r_i}}{\partial{x_1}\partial{x_n}}       \\  
  \frac{\partial^2{r_i}}{\partial{x_2}\partial{x_1}} & \frac{\partial^2{r_i}}{\partial{x_2}\partial{x_2}}  & \cdots & \frac{\partial{r_i}}{\partial{x_2}\partial{x_n}}       \\ 
 \vdots & \vdots  & \ddots & \vdots\\  
 \frac{\partial^2{r_i}}{\partial{x_n}\partial{x_1}} & \frac{\partial^2{r_i}}{\partial{x_n}\partial{x_2}}  & \cdots & \frac{\partial{r_i}}{\partial{x_n}\partial{x_n}}      \\  
\end{bmatrix} $$
$$

In 3d points matching problem, the increment of residual function is shown below:  
$$ r = T \exp{(\delta x)}a - b $$
* guass_newton_method.md (5)(6)(7)

```python
r = T0*expx*a-b
r = Matrix(r[0:3])
print("Our residual vector is show in guass_newton_method.md (7)")
display(Math(r'r = %s' % (sympy.latex(r))))
```

```python
print("Use the definition of Jacobian:")
dr_real = Matrix([[diff(r[0],x1),diff(r[0],x2),diff(r[0],x3),diff(r[0],x4),diff(r[0],x5),diff(r[0],x6)],
                  [diff(r[1],x1),diff(r[1],x2),diff(r[1],x3),diff(r[1],x4),diff(r[1],x5),diff(r[1],x6)],
                  [diff(r[2],x1),diff(r[2],x2),diff(r[2],x3),diff(r[2],x4),diff(r[2],x5),diff(r[2],x6)]])
display(Math(r'\dot{r} = %s' % (sympy.latex(dr_real))))
print("Use guass_newton_method.md (12):")
dr = Matrix.hstack(R0,-R0*skew(a))
display(Math(r'\dot{r} = %s' % (sympy.latex(dr))))


if(dr.expand()==dr_real.expand()):
    print("The calculation results are exactly the same!")
else:
    print("The calculation results are different!")
```

```python
print("Our objective function is show in newton_method.md(2)")
#newton_method.md (2)
f=r.T*Sigma*r/2
f = f.expand()

display(Math(r'f = %s' % (sympy.latex(f))))
```

```python
df_real = Matrix([diff(f[0],x1),diff(f[0],x2),diff(f[0],x3),diff(f[0],x4),diff(f[0],x5),diff(f[0],x6)]).T
df = r.T*Sigma*dr
df_real = df_real.expand()
df = df.expand()

print("Use the definition of Jacobian:")
display(Math(r'\dot{f} = %s' % (sympy.latex(df_real))))
print("Use newton_method.md (6):")
#newton_method.md (6)
display(Math(r'\dot{f} = %s' % (sympy.latex(df))))

if(df==df_real):
    print("The calculation results are exactly the same!")
else:
    print("The calculation results are different!")
```

```python
d2f_real = Matrix([[diff(f[0],x1,x1),diff(f[0],x1,x2),diff(f[0],x1,x3),diff(f[0],x1,x4),diff(f[0],x1,x5),diff(f[0],x1,x6)],
                   [diff(f[0],x2,x1),diff(f[0],x2,x2),diff(f[0],x2,x3),diff(f[0],x2,x4),diff(f[0],x2,x5),diff(f[0],x2,x6)],
                   [diff(f[0],x3,x1),diff(f[0],x3,x2),diff(f[0],x3,x3),diff(f[0],x3,x4),diff(f[0],x3,x5),diff(f[0],x3,x6)],
                   [diff(f[0],x4,x1),diff(f[0],x4,x2),diff(f[0],x4,x3),diff(f[0],x4,x4),diff(f[0],x4,x5),diff(f[0],x4,x6)],
                   [diff(f[0],x5,x1),diff(f[0],x5,x2),diff(f[0],x5,x3),diff(f[0],x5,x4),diff(f[0],x5,x5),diff(f[0],x5,x6)],
                   [diff(f[0],x6,x1),diff(f[0],x6,x2),diff(f[0],x6,x3),diff(f[0],x6,x4),diff(f[0],x6,x5),diff(f[0],x6,x6)]])

d2f = dr.T * Sigma * dr 

d2f_real = d2f_real.expand()
d2f = d2f.expand()
print("Use the definition of Hessian:")
display(Math(r'\ddot{f} = %s' % (sympy.latex(d2f_real))))
print("Use newton_method.md (11):")
display(Math(r'\ddot{f} = %s' % (sympy.latex(d2f))))

if(d2f==d2f_real):
    print("The calculation results are exactly the same!")
else:
    print("The calculation results are different!")
```
